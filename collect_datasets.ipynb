### Convert to AnnData

```python

import anndata as ad

# Step 1: Load your tabular data (e.g., from CSV)
df = fdpd.read_csv("data/pangolin/pangolin_heart_celltype.csv", sep="\t")  # or pd.read_table("your_data.tsv")

df = df.drop(columns=["Project"])
df
```

```python
# Step 2: Extract expression matrix (gene expression values)
X = df.drop(columns=["Celltype", "Cells"]).values  # numpy array of expression data

# Step 3: Create observation (cell-level) metadata
obs = pd.DataFrame({"Celltype": df["Celltype"]})
obs.index = df["Cells"] #[f"cell_{i}" for i in range(len(df))]  # optional: name cells

# Step 4: Create variable (gene-level) metadata
var = pd.DataFrame(index=df.columns.drop(["Celltype", "Cells"]))  # gene names as index

# Step 5: Create AnnData object
adata = ad.AnnData(X=X, obs=obs, var=var)

# (Optional) Save to file
adata.write("data/pangolin/pangolin_heart.h5ad")
```

```python
from sklearn.model_selection import train_test_split
import anndata as ad

# Split indices
train_idx, test_idx = train_test_split(
    range(adata.n_obs), test_size=0.2, random_state=42, stratify=adata.obs["Celltype"]
)

# Create new AnnData objects
adata_train = adata[train_idx].copy()
adata_test = adata[test_idx].copy()

# Optionally save to disk
adata_train.write("data/pangolin/pangoling_heart_train.h5ad")
adata_test.write("data/pangolin/pangoling_heart_test.h5ad")
```
